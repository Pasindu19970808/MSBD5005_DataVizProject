```python
import numpy as np
import pandas as pd
import geopandas as gpd
import altair as alt
from preprocessing import *
from pathlib import Path
```

```python
building_alt_data = plottingUtils.processBuildingData('building_location_col')
```

```python
data_apartments = plottingUtils.processApartmentParticipantDetails('id_col','apartments')
data_apartments = plottingUtils.addHappinessScore(data_apartments)
```

```python
data_apartments['x'] = data_apartments['x'].astype(float)
data_apartments['y'] = data_apartments['y'].astype(float)
```

```python
#plot buildings
background = alt.Chart(building_alt_data).mark_geoshape(
    opacity = 0.75
).encode(
    color = 'properties.buildingType:N'
).project(
    type='identity', reflectY=True
).properties(
    width = 750,
    height = 750
)

joviality = alt.Chart(data_apartments).mark_circle(size = 100).encode(
    latitude = 'y',longitude = 'x',
    color = alt.Color('joviality',scale = alt.Scale(
    domain = list(np.linspace(data_apartments["joviality"].min(),data_apartments['joviality'].max(),3)),
    # domain = [-10,-5,0,5,9],
    range = ['red','yellow','green'],
    interpolate = 'rgb'
    ),
    legend=alt.Legend(direction='vertical', orient='right', title='Happiness'))).properties(
    width = 750,
    height = 750
)
# alt.layer(background,joviality).resolve_scale(x = 'shared',y = 'shared')
background + joviality
```
