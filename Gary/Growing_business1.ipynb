```python
import numpy as np
journal_base_path = jornal_base_path

checkin_df = pd.read_csv(journal_base_path / 'CheckinJournal.csv')
```

```python
checkin_df['datetime'] = pd.to_datetime(checkin_df['timestamp'], format='%Y-%m-%dT%H:%M:%SZ')
checkin_df['date'] = checkin_df['datetime'].dt.date
```

```python
travel_df['checkin_date'] = travel_df['checkin_datetime'].dt.date
travel_checkin_df = travel_df[['checkin_date', 'travelEndLocationId', 'purpose']].groupby(by=['checkin_date', 'travelEndLocationId', 'purpose']).size().reset_index(drop=False)
venue_daily_checkin_df = checkin_df[['venueId', 'venueType', 'date']].groupby(by=['venueId', 'venueType', 'date']).size().reset_index(drop=False)
venue_daily_checkin_df.columns =['venueId', 'venueType', 'checkin_date', 'checkin_cnt']
travel_checkin_df.columns = ['checkin_date', 'venueId', 'purpose', 'checkin_cnt_travel']
daily_venue_checkin = travel_checkin_df.merge(venue_daily_checkin_df, on=['checkin_date', 'venueId'], how='left')
daily_venue_checkin['checkin_yrwk'] = daily_venue_checkin['checkin_date'].apply(lambda x: x.strftime("%G%V"))
daily_venue_checkin['weekday'] = daily_venue_checkin['checkin_date'].apply(lambda x: x.weekday())

travel_df['cost'] = travel_df[['startingBalance', 'endingBalance']].apply(lambda x: x[0] - x[1], axis=1)
travel_df['yrwk'] = travel_df['checkin_date'].apply(lambda x: x.strftime("%G%V"))

venue_daily_spend = travel_df[['checkin_date', 'cost', 'travelEndLocationId', 'purpose']].groupby(by=['checkin_date', 'travelEndLocationId','purpose']).agg(
    income_mean=('cost', np.mean),
    income_sum=('cost', np.sum),
    income_max=('cost', np.max),
    income_min=('cost', np.min),
).reset_index(drop=False)
venue_daily_spend.columns = ['checkin_date', 'venueId', 'purpose', 'incomce_mean', 'income_sum', 'income_max', 'income_min']
daily_venue_checkin_income = daily_venue_checkin.merge(venue_daily_spend, on=['checkin_date', 'venueId', 'purpose'], how='left')
daily_venue_checkin_income['checkin_yrwk'] = daily_venue_checkin_income['checkin_date'].apply(lambda x: x.strftime('%G%V'))
```

```python
daily_venue_checkin_income.to_csv('daily_venue_checkin_income.csv', index=False)
```

```python
daily_venue_checkin_income['yr_month'] = daily_venue_checkin_income['checkin_date'].apply(lambda x: x.strftime("%Y-%m"))
```

```python
daily_venue_checkin_income
```

```python
monthly_venue_checkin_income = daily_venue_checkin_income[['venueId', 'purpose', 'checkin_cnt_travel', 'venueType', 'yr_month', 'income_sum', 'incomce_mean', 'Occupancy_ratio']].groupby(
    ['venueId', 'purpose', 'venueType', 'yr_month']
).agg(
    monthly_customer_count=('checkin_cnt_travel', np.sum),
    monthly_income=('income_sum', np.sum),
    customer_average_spend=('incomce_mean', np.mean),
    monthly_occupancy_ratio=('Occupancy_ratio', np.sum)
).reset_index()
```

```python
monthly_venue_checkin_income =  monthly_venue_checkin_income.sort_values(by=['venueId', 'purpose', 'yr_month'])
```

```python
monthly_venue_checkin_income['last_month_income'] = monthly_venue_checkin_income.groupby(by=['venueId', 'purpose', 'venueType'])['monthly_income'].shift(1)
monthly_venue_checkin_income['last_month_customer_count'] = monthly_venue_checkin_income.groupby(by=['venueId', 'purpose', 'venueType'])['monthly_customer_count'].shift(1)
monthly_venue_checkin_income['last_month_occupancy_ratio'] = monthly_venue_checkin_income.groupby(by=['venueId', 'purpose', 'venueType'])['monthly_occupancy_ratio'].shift(1)
monthly_venue_checkin_income['last_month_customer_average_spend'] = monthly_venue_checkin_income.groupby(by=['venueId', 'purpose', 'venueType'])['customer_average_spend'].shift(1)
```

```python
monthly_venue_checkin_income['last_month_customer_average_spend'] = monthly_venue_checkin_income['last_month_customer_average_spend'].fillna(0)
```

```python
monthly_venue_checkin_income.to_csv('monthly_venue_checkin_income.csv')
```
