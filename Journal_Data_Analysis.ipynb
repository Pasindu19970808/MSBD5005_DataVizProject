## Financial Journal


```python
financejournal = pd.read_csv("./Datasets/Journals/FinancialJournal.csv")
financejournal = financejournal.sort_values(by=["participantId", "timestamp"])
```

```python
financejournal["rec_date"] = financejournal["timestamp"].str[:10]
daily_financejournal = (
    financejournal.groupby(by=["participantId", "rec_date", "category"])["amount"]
    .sum()
    .reset_index()
)
daily_financejournal.to_csv(
    "./Datasets/Journals/Daily_FinancialJournal.csv", index=False
)
print('saved daily finance journal........')
```

```python

daily_financejournal["yr_mth"] = daily_financejournal["rec_date"].str[:7]
monthly_financejournal_details = (
    daily_financejournal.groupby(by=["participantId", "yr_mth", "category"])["amount"]
    .sum()
    .reset_index()
)
monthly_financejournal_details = monthly_financejournal_details.pivot_table(
    index=["participantId", "yr_mth"],
    columns=["category"],
    values=["amount"],
    fill_value=0,
).reset_index()

monthly_financejournal_details.columns = ['participantId', 'yr_mth'] + monthly_financejournal_details.reset_index().columns.droplevel().tolist()[3:]
monthly_financejournal_details['NetBalance'] = monthly_financejournal_details.iloc[:,2:].sum(axis=1)
monthly_financejournal_details.to_csv(
    "./Datasets/Journals/Monthly_FinanceJournal_Details.csv", index=False
)
print('saved monthy finance journal details.........')
```
