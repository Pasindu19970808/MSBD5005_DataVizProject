```python
import pandas as pd
import numpy as np
```

```python
ALL_ACT_LOG_FILE = "all_act_log.csv"
JOBS_FILE = "./Datasets/Attributes/Jobs.csv"
```

```python
all_act_log_df = pd.read_csv(ALL_ACT_LOG_FILE)
jobs_df = pd.read_csv(JOBS_FILE)
```

```python
all_act_log_df = all_act_log_df.drop(columns=['x','y'])
```

```python
all_act_log = all_act_log_df.merge(
    right=jobs_df,
    how='left',
    on=['jobId']
)
```

```python
all_act_log.head()
```

```python
all_act_log['date'] = all_act_log['timestamp'].str[:10]
```

```python
df = []
for i in range(1, 73):
    df.append(pd.read_csv(f'./Datasets/Activity Logs/ParticipantStatusLogs{i}.csv'))
df = pd.concat(df, axis = 0)
```

```python
for i in set(df['participantId']):
    df[df['participantId'] == i].sort_values(by=['timestamp']).to_csv(f'participant_{i}_all_activity.csv', index=False)
    print('saved', i)
```
